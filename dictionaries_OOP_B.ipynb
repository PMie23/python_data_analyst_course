```python
file = open('mbox.txt', 'r')
email_dict = {}
for f in file:
    if f.startswith('From '):
        key = f.split()[1]
        if key in email_dict:
            email_dict[key] += 1
        else:
            email_dict[key] = 1
file.close()
```

```python
file = open('mbox.txt', 'r')
domain_dict = {}
for f in file:
    if f.startswith('From:'):
        index = f.find('@')
        domain_name = f[index+1:].replace('\n','')
        if domain_name in domain_dict.keys():
            domain_dict[domain_name] = domain_dict[domain_name]+1
        else:
            domain_dict[domain_name] = 1
file.close()
```

```python
class Shapes():          # Base class
    def __init__(self):
        self.length = 1
    def area(self):
        pass
    def perimeter(self):
        pass             # Return NotImplementedError
```

```python
class Circle(Shapes):    # inheritance
    def area(self):
        return 3.14*self.length*self.length
    def perimeter(self):
        return 2*3.14*self.length
```

```python
circle1 = Circle()
```

```python
circle1.length = 10
```

```python
circle1.area()
```

```python
circle1.perimeter()
```

```python
circle2 = Circle()
circle2.length = 25
circle2.area()
```

```python
class Rectangle(Shapes):
    def __init__(self):
        self.length2 = 1
    def area(self):
        return self.length*self.length2
    def perimeter(self):
        return 2*(self.length+self.length2)
```

```python
rectangle1 = Rectangle()
```

```python
rectangle1.length = 25
rectangle1.length2 = 10
rectangle1.area()
```

```python
rectangle1.perimeter()
```
